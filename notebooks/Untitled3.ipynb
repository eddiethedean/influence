```python
df = pd.read_csv('../data/tcc_ceds_music.csv').drop('Unnamed: 0', axis=1)
df
```

```python
dists = pd.read_csv('../data/dists.csv')
dists
```

```python
def get_top_ten(df, dists, x_index):
    f = dists['x_index'] == x_index
    indexes = list(dists[f]['y_index'])
    return df.iloc[indexes]
```

```python
get_top_ten(df, dists, 28370)
```
